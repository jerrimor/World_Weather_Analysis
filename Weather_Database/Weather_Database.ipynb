```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Import the time library and the datetime module from the datetime library 
import time
from datetime import datetime
```

```python
#create set of 2,000 random latitude and longitude
lats = np.random.uniform(low = -90.000, high = 90.000, size = 2000)
lngs = np.random.uniform(low = -180.000, high = 180.000, size = 2000 )
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
#create a list of lat, lngs from zip file
coordinates = list(lat_lngs)

```

```python
#get nearest city

#import citipy
from citipy import citipy

#create a list for the cities
cities = []
#identify the nearest city for each lat/lon pair
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
#if the city is unique we can add it to the city list
    if city not in cities:
        cities.append(city)
#print to verify enough cities were generated
len(cities)
```

```python
#api call with OpenWeatherMap

#import requests
import requests
from config import weather_api_key

#build WeatherPy URL
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key



```

```python
# Create an empty list to hold the weather data.
city_data = []
# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters.
record_count = 1
set_count = 1


#loop through cities in our list
for i, city in enumerate(cities):
    
    #group cities in sets of 50
    if (i % 50 ==0 and i >= 50):
        set_count += 1
        record_count = 1
        time.sleep(60)
    
    #create endpoint URL with each city
    city_url = url + "&q=" + city.replace(" ", "+")
    #print(city_url)
    
    #log the url, record and set the numbers and city
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    #add 1 to the record count.
    record_count += 1    
    
    try:

        city_weather = requests.get(city_url).json()

        #info to parse out of API call
    
        lat = city_weather["coord"]["lat"]
        lng = city_weather["coord"]["lon"]
        city_title = city_weather["name"]
        max_temp = city_weather["main"]["temp_max"]
        humidity = city_weather["main"]["humidity"]
        clouds = city_weather["clouds"]["all"]
        wind = city_weather["wind"]["speed"]
        weather_descrip = city_weather["weather"][0]["description"]
        country = city_weather["sys"]["country"]
        #print(lat, lng, max_temp, humidity, clouds, wind, weather_descrip)
    
        #adding this data to the city dataframe
        city_data.append({"City" : city_title,
                          "Country" : country,
                         "Lat" : lat,
                         "Lng" : lng,
                         "Max Temp" : max_temp,
                         "Humidity" : humidity,
                          "Cloudiness" : clouds,
                          "Wind Speed" : wind,
                          "Weather Description" : weather_descrip})
        
        #error handling
    except:
            print('City not found.  Skipping...')
            pass
                  
#indicate that data loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
#confirm data is within city_data 
len(city_data)
```

```python
#create DF from json for city weather data
city_weather_df = pd.DataFrame(city_data)
city_weather_df.head()
```

```python
#create csv from DF
output_data_file = 'WeatherPy_Database.csv'
#export the data
city_weather_df.to_csv(output_data_file, index_label="City_ID")
```
